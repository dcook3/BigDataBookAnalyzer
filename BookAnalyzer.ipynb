```python
import re
with open(r"BigData.txt") as f:
    #Read each line of file and add into list
    lines = f.readlines()

words = []

#Track line index as well as line
i = 0
for line in lines:
    #Split all words on space (default)
    split_line = line.split()
    
    #Find non-desired characters make new word without them and replace old word with new in same postion
    for i in range(len(split_line)):
        #Remove any occurance of non-alphabetical characters by removing matches using this regular expression
        newword = re.sub(r"[^\w]|\d", "", split_line[i])
        split_line[i] = newword
    
    #Right now, split line is still not in its most simple form. Each line is represented by a list.
    #We would rather have one list with all words, not a 2 dimensional list of lines with words.
    for word in split_line:
        words.append(word.lower())    
    i += 1
    
print(words)
```

```python
def value_getter(item):
    return item[1]

#Track words using a word:wordcount dictionary.
words_dict = dict()

for word in words:
    #Determine whether exists to know whether to add a new k/v pair or increment the matching key word
    if word not in words_dict:
        words_dict[word] = 1
    else:
        words_dict[word] = words_dict[word] + 1

#remove '' from words_dict
words_dict.pop('')
#Sort using sorted(), in this case we pull out word and use value_getter to maintain k/v pairs during sort.
words_sorted = dict(sorted(words_dict.items(), key=value_getter, reverse=True))
        
print(words_sorted)
```

```python
print(len(words_dict))
```

```python
#top words from words_sorted
top_words = list(words_sorted.items())[:100]
print(top_words)
```

```python
#calculate the percentage of the original document that each word makes up
num_words = len(words)
in_text_frequency = dict()
for word in top_words:
    word_count = word[1]
    percentage = word_count / num_words
    in_text_frequency[word[0]] = percentage
print(in_text_frequency)
    
```

```python
#get the frequency of the top words in the english language using datamuse API
import requests
import json

top_words_frequency = dict()
for word in top_words:
    word_frequency_request = requests.get("https://api.datamuse.com/words?sp=" + word[0] + "&md=f")
    word_frequency_json = word_frequency_request.json()
    word_frequency = word_frequency_json[0]['tags'][0][2:]
    #f is occurances per million, so divide by 1000000 to get frequency
    word_frequency = float(word_frequency) / 1000000
    top_words_frequency[word[0]] = word_frequency
print(top_words_frequency)
```

```python
percent_gap = dict()
for word in top_words:
    percent_gap[word[0]] = in_text_frequency[word[0]]/top_words_frequency[word[0]]
percent_gap_sorted = dict(sorted(percent_gap.items(), key=value_getter, reverse=True))
print(percent_gap_sorted)
```
